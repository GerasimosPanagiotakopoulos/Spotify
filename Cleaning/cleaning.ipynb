```python
data_full.drop('Index', axis=1, inplace=True)
data_full.head()
```

```python
data_full['Release Date'] = pd.to_datetime(data_full['Release Date'], format='%Y-%m-%d', errors='coerce')
```

```python
int_columns = ['Streams', 'Artist Followers', 'Popularity', 'Duration (ms)']

data_full['Streams'] = data_full['Streams'].str.replace(',', '')
data_full['Artist Followers'] = data_full['Artist Followers'].str.replace(' ', '0')
data_full['Popularity'] = data_full['Popularity'].str.replace(' ', '0')
data_full['Duration (ms)'] = data_full['Duration (ms)'].str.replace(' ', '0')

for col in int_columns:
    data_full[col] = data_full[col].astype('int64')
```

```python
numeric_columns = ['Danceability', 'Energy', 'Loudness', 'Speechiness', 'Acousticness', 'Liveness', 'Tempo', 'Valence']
for col in numeric_columns:
    data_full[col] = pd.to_numeric(data_full[col], errors='coerce')
```

```python
data_full['Lead Artist'] = data_full['Artist'].str.split(',').str[0]
```

```python
data_full['Date of Highest Charting'] = data_full['Week of Highest Charting'].str.split('--').str[0]
data_full['Date of Highest Charting'] = pd.to_datetime(data_full['Date of Highest Charting'], format='%Y-%m-%d')
```

```python
data_full.to_csv('../data/spotify_dataset_clean.csv', index=False)
```
